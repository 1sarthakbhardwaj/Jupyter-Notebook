```python
from matplotlib import pylab 
print(pylab.__version__)
```

```python
import numpy as np
x = np.linspace(1,9,40)
y = x *x+ 2
print(x)
print(y)
```

```python
pylab.plot(x,y ,"b")
```

```python
from matplotlib import pylab as plt
fig = plt.figure()
axis = fig.add_axes([1,2,1.1,0.2])
axis.plot(x,y,"r")
```

```python
fig, axes = plt.subplots(nrows = 1 , ncols = 4)
for ax in axes:
    ax.plot(x,y,"b")
```

```python
fig = plt.figure()
#add_axes parameter contains left , bottom,width,height

axis1 = fig.add_axes([0.1,0.1,0.8,0.8])
axis2 = fig.add_axes([0.1,0.1,0.1,0.1])

axis1.plot(x,y,"r")
axis2.plot(x,y,"b")
```

```python
fig, axes = plt.subplots()
axes.set_title("title") #adding title
axes.set_xlabel("x") # adding labels
axes.set_ylabel("y")
axes.plot(x,y,"g--")
```

# legend

```python

fig, axes = plt.subplots()
axes.set_title("legend")
axes.set_xlabel("x label")
axes.set_ylabel("y label")
axes.plot(x,x**3,"r")
axes.plot(x,x**2,"b")
axes.legend(["x_label","y_label"],loc = 0)
```

```python
#changing size of properties

fig, axes = plt.subplots(dpi = 60)

axes.plot(x,x+1,"r", alpha = .8)
axes.plot(x,x+2,"b",alpha = .3)
axes.plot(x,x+3,"g--", alpha = .10)
```

```python
fig ,ax = plt.subplots()
ax.plot(x,x+2,color = "blue" , lw = 3, linestyle = "-.",marker ="o")
ax.plot(x,x+2,color = "blue" , lw = 3, linestyle = "-.")
ax.plot(x,x+2,color = "blue" , lw = 3, linestyle = "-.")
ax.plot(x,x+2,color = "blue" , lw = 3, linestyle = "-.")
```
