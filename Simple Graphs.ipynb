# different types of graphs

```python
from matplotlib import pylab as plt
fig = plt.fig()
import numpy as np
```

```python
x = np.linspace(1,40,14)
y = x *2
```

```python
fig ,axes = plt.subplots(1,2, figsize = (10,5))
axes[0].plot(x,x+2 ,x,x+5 ,"r", "r",lw = 2)
axes[0].grid(True)
axes[1].plot(x,x+4,x,x+1,x,x+6,"r")
axes[1].set_xlim([0,6])
axes[1].set_ylim([0,6])
```

## 2D Graphs

```python
fig, axes = plt.subplots(1,4, figsize = (16,5))
axes[0].set_title("scatter")
axes[0].scatter(x,x**2,lw = "1.3")
axes[1].set_title("bar")
axes[1].bar(x,x**2,lw = "1.3")
axes[2].set_title("hist")
axes[2].hist(x,x**2,lw = "1.3")
axes[2].set_xlim([0,400])
axes[2].set_ylim([0,40])
axes[3].set_title("step")
axes[3].step(x,x**2,lw = "1.3")
```

```python
r = np.random.randn(1000)
fig ,ax = plt.subplots(1,2,figsize = (15,5))
ax[0].hist(r)
ax[1].hist(r, cumulative =True , bins = 39)
```
